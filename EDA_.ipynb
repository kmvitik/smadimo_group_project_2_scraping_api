```python
import pandas as pd
import matplotlib.pyplot as plt

table = pd.read_csv("vacancies_cleaned.csv")
table = table.drop_duplicates(subset=["link"])
table.shape
```

```python
employment_counts = table["employment_type"].value_counts()
employment_counts
```

```python
plt.figure(figsize=(16, 9))
plt.bar(employment_counts.index, employment_counts.values)
plt.title("Количество вакансий по времени занятости")
plt.xlabel("время-занятости")
plt.ylabel("кл-во вакансий")
plt.show()
```

```python
best_locations = table["location"].value_counts().head(10)
best_locations
```

```python
plt.figure(figsize=(16, 9))
plt.bar(best_locations.index, best_locations.values)
plt.title("10 локаций, где больше всего вакансий")
plt.xlabel("локация")
plt.ylabel("к-во вакансий")
plt.show()
```

```python
how_many_each_grades = (
    table[
        ["trainee", "junior", "middle", "senior", "lead", "head", "director", "c_level"]
    ]
    .sum()
    .sort_values(ascending=False)
)
how_many_each_grades
```

```python
plt.figure(figsize=(16, 9))
plt.bar(how_many_each_grades.index, how_many_each_grades.values)
plt.title("Распрделение вакансий по уровням")
plt.xlabel("Уровень")
plt.ylabel("Количество вакансий")
plt.show()
```

```python
table["company"] = table["company"].str.strip()
table["company"] = table["company"].replace({"Yandex": "Яндекс"})
```

```python
ok_companies = table[table["company"] != "Company hidden"]
best_companies = ok_companies["company"].value_counts().head(10)
best_companies
```

```python
plt.figure(figsize=(16, 9))
plt.bar(best_companies.index, best_companies.values)
plt.title("10 компаний, где есть больше всего вакансий")
plt.xlabel("компания")
plt.ylabel("Кол-во вакансий")
plt.xticks(rotation=55)
plt.show()
```

```python
salary_table = table.dropna(subset=["salary"])
salary_locations = salary_table["location"].value_counts().head(5)
salary_locations
```

```python
plt.figure(figsize=(16, 9))
plt.bar(salary_locations.index, salary_locations.values)
plt.title(
    "5 стран-лидеров с вакансиями, где открыты зарплатные ожидания для кандидатов"
)
plt.xlabel("Страна")
plt.ylabel("кл-во вакансий")
plt.show()
```
