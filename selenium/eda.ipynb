```python
def str_to_list(s):
    if s.startswith("[") and s.endswith("]"):
        s = s[1:-1]
        ss = s.split(", ")
        ss = [ss.replace("'", "") for ss in ss]
        ss = [ss.replace(" ", "") for ss in ss]
        return ss
    else:
        return []


str_to_list(str_to_parse)
```

Вкусняшка так называемая мало того что в вакансии могут юыть сразу несколько типов так еще и страну в () храним, идея следующая сначала попаем скобки а инфу из них кладем в новый столбец потом по / сплитим и получается у нас 3 бинарных столбца (one hot encoding) onsite, hybrid, remote + еще столбец со страной.

```python
def parse_work_type(l):
    country = None
    answer = {
        "onsite": False,
        "remote": False,
        "hybrid": False,
        "country": None,
        "tags": [],
    }
    parsed_tag = None
    for tag in l:
        if (
            tag.find("onsite") != -1
            or tag.find("remote") != -1
            or tag.find("hybrid") != -1
        ):
            s = tag.find("(")
            e = tag.find(")")
            if s != -1 and e != -1:
                country = tag[s + 1 : e]
                new_tag = tag[:s] + tag[e + 1 :]
                print(country, new_tag)
            else:
                country = None
                new_tag = tag
            if tag.find("onsite"):
                answer["onsite"] = True
            if tag.find("remote"):
                answer["remote"] = True
            if tag.find("hybrid"):
                answer["hybrid"] = True
            answer["country"] = country
            parsed_tag = tag
    answer["tags"] = [tag for tag in l if tag != parsed_tag]
    return answer


parse_work_type(df["tags_parsed"].iloc[5])
```

```python
tags = df["tags_parsed"]
for i in range(len(tags)):
    parsed = parse_work_type(tags[i])
    df.at[i, "onsite"] = parsed["onsite"]
    df.at[i, "remote"] = parsed["remote"]
    df.at[i, "hybrid"] = parsed["hybrid"]
    df.at[i, "country"] = parsed["country"]
    df.at[i, "tags_parsed"] = parsed["tags"]
df.head(20)
```

```python
def parse_level_type(l):
    answer = {
        "tags": l,
        "lead": False,
        "c_level": False,
        "senior": False,
        "director": False,
        "trainee": False,
        "head": False,
        "middle": False,
        "junior": False,
    }
    parsed_tag = None
    for tag in l:
        if tag.find("lead") != -1:
            answer["lead"] = True
            parsed_tag = tag
        if tag.find("c_level") != -1:
            answer["c_level"] = True
            parsed_tag = tag
        if tag.find("senior") != -1:
            answer["senior"] = True
            parsed_tag = tag
        if tag.find("director") != -1:
            answer["director"] = True
            parsed_tag = tag
        if tag.find("trainee") != -1:
            answer["trainee"] = True
            parsed_tag = tag
        if tag.find("head") != -1:
            answer["head"] = True
            parsed_tag = tag
        if tag.find("middle") != -1:
            answer["middle"] = True
            parsed_tag = tag
        if tag.find("junior") != -1:
            answer["junior"] = True
            parsed_tag = tag
    answer["tags"] = [tag for tag in l if tag != parsed_tag]
    return answer


parse_level_type(df["tags_parsed"].iloc[5])
```

```python
tags = df["tags_parsed"]
for i in range(len(tags)):
    parsed = parse_level_type(tags[i])
    df.at[i, "lead"] = parsed["lead"]
    df.at[i, "c_level"] = parsed["c_level"]
    df.at[i, "senior"] = parsed["senior"]
    df.at[i, "director"] = parsed["director"]
    df.at[i, "trainee"] = parsed["trainee"]
    df.at[i, "head"] = parsed["head"]
    df.at[i, "middle"] = parsed["middle"]
    df.at[i, "junior"] = parsed["junior"]
    df.at[i, "tags_parsed"] = parsed["tags"]
df.head(20)
```
