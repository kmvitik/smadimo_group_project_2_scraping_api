```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('vacancies.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

Итак надо поработать с тегами

```python
print(type(df['tags'].iloc[0]))
str_to_parse = df['tags'].iloc[0]
print(str_to_parse)
```

```python
def str_to_list(s):
    if s.startswith('[') and s.endswith(']'):
        s = s[1:-1]
        ss = s.split(', ')
        ss = [ss.replace("'", "") for ss in ss]
        ss = [ss.replace(' ', '') for ss in ss]
        return ss
    else:
        return []

str_to_list(str_to_parse)
```

Имба функция вроде работает попробуем ее применить

```python
df['tags_parsed'] = df['tags'].apply(lambda x: str_to_list(x))
print(type(df['tags_parsed'].iloc[0]))
df.head()
```

```python
df['len_tags']= df['tags_parsed'].apply(lambda x: len(x))
df.head()
```

```python
df['len_tags'].value_counts()
```

Я изначально запутался в числе возможных тегов, их оказывается 5, а не 4, поэтому проведя анализ по 5 тегам, видим, что одно из них это relocation, соответственно просто превращаем его в флаг, есть ли relocation или нет. Для всего остального будем исследовать датасет уже с 4 тегами.

Так, с relocation вроде разобрались. Теперь уберем ее из исходного столбца tags_parsed.

```python
def del_tag(l, tag):
    if tag in l:
        l.remove(tag)
    return l

df['tags_parsed'] = df['tags_parsed'].apply(lambda x: del_tag(x, 'Relocation'))
```

```python
df['len_tags']= df['tags_parsed'].apply(lambda x: len(x))
df['len_tags'].value_counts()
```

Ура! Уже видно какое-то движение. Теперь у нас максимум 4 тега.

Ну все, нигде больше он не встречается Соответственно точно заменяем эту штуку

```python
def replace_tag(l, tags):
    for tag in tags:
        if tag in l:
            l.remove(tag)
            return tag
    return None
```

```python
employment_types = ['fulltime', 'parttime', 'project']
df['employment_type'] = df['tags_parsed'].apply(lambda x: replace_tag(x, employment_types))
print(df['employment_type'].isna().sum())
df['employment_type'].value_counts()
```

```python
for tag in employment_types:    
    df['tags_parsed'] = df['tags_parsed'].apply(lambda x: del_tag(x, tag))
df['len_tags']= df['tags_parsed'].apply(lambda x: len(x))
df['len_tags'].value_counts()
```

Ура! Победа! Еще один тип tag спаршен

```python
skill_all = pd.DataFrame()
skill_all['list'] = df[df['len_tags'] > 2]['tags_parsed']
skill_all['1_skill'] = skill_all['list'].apply(lambda x: x[0])
skill_all['2_skill'] = skill_all['list'].apply(lambda x: x[1])
skill_all['3_skill'] = skill_all['list'].apply(lambda x: x[2])
skill_all.head()
```
