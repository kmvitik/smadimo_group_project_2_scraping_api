Данные в колонке представлены строковым типом. для дальнейшей обработки преобразуем в массив с помощью функции: 

```python
data['skills'].sample(3)
```

```python
def str_to_list(skills_string):
    if skills_string == "[]":
        return []
    skills_string = skills_string[1:-1].split(",")
    skills_list = [skill.replace("'", "").strip() for skill in skills_string]
    return skills_list
```

```python
sample1 = "[]"
sample2 = "['typescript', 'react', 'angular', 'git', 'linux', 'mvvm', 'fsd', 'ci/cd', 'тестирование', 'архитектура']"
str_to_list(sample1)
```
