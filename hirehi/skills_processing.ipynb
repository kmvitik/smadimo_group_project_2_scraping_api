В этом ноутбуке формируется дополнительная таблица skills.
- Строка таблицы - `id` вакансии и соответствующий навык `skill`
- Связь с вакансиями из исходной таблицы осуществляется через `id`

```python
data = pd.read_csv('vacancys_final.csv')
data.sample(3)
```

```python
data['skills'].value_counts()
```

Данные в колонке представлены строковым типом. для дальнейшей обработки преобразуем в массив с помощью функции: 

```python
data['skills'].sample(3)
```

```python
def str_to_list(skills_string):
    if skills_string == "[]":
        return []
    skills_string = skills_string[1:-1].split(",")
    skills_list = [skill.replace("'", "").strip() for skill in skills_string]
    return skills_list
```

Для формирования дополнительного столбца используется метод [`explode`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.explode.html):

```python
data["skills"] = data["skills"].apply(lambda x: str_to_list(x))
```

```python
data_skills = data[["id", "skills"]].explode("skills").rename(columns={'skills': 'skill'})
data_skills
```
