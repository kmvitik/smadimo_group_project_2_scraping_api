нужно строки со словарями обработать как словари https://docs.python.org/3/library/ast.html#ast.literal_eval

```python
import ast
```

```python
df["location"] = df["location"].apply(ast.literal_eval)
df["company"] = df["company"].apply(ast.literal_eval)
```

```python
df["location"] = df["location"].str["display_name"]
df["company"] = df["company"].str["display_name"]
```

```python
df[["title", "location", "company"]].head()
```

```python
df.info()
```

```python
df["salary_avg"] = df[["salary_min", "salary_max"]].mean(axis=1)
```

```python
import matplotlib.pyplot as plt
```

```python
df_plot = df[["salary_avg", "location"]].head(20)
df_plot = df_plot.sort_values("salary_avg")
```

```python
plt.figure(figsize=(16, 9))
plt.barh(df_plot["location"], df_plot["salary_avg"])


plt.title("топ 20 регионов Великобритании по средней зарплате", fontsize=20)
plt.xlabel("средняя зарплата (фунты)", fontsize=16)
plt.ylabel("регион", fontsize=16)
plt.grid(axis="y", color="black", linestyle="-.", linewidth=0.3)

plt.show()
```

```python
df_plot = df.sort_values("salary_avg", ascending=False).head(20)
```

```python
df_plot = (
    df.groupby("company", as_index=False)["salary_avg"]
    .mean()
    .sort_values("salary_avg", ascending=False)
    .head(20)
)

plt.figure(figsize=(16, 10))
plt.barh(df_plot["company"], df_plot["salary_avg"], color="steelblue")
plt.gca().invert_yaxis()

plt.title("топ 20 компаний по средней зарплате", fontsize=20)
plt.xlabel("средняя зарплата (фунты)", fontsize=16)
plt.ylabel("компания", fontsize=16)
plt.grid(axis="y", color="black", linestyle="-.", linewidth=0.3)

plt.show()
```
